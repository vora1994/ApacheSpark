```
from pyspark.sql import SparkSession
from graphframes import *
import sql
```

```
GBvideos = sqlContext.sql("SELECT * FROM GBvideos_csv")
GBcomments =sqlContext.sql("SELECT * FROM GBcomments_csv")
USvideos = sqlContext.sql("SELECT * FROM USvideos_csv")
UScomments =sqlContext.sql("SELECT * FROM UScomments_csv")
```

```
%sql
select * from GBvideos_csv
```

```
%sql
select * from GBcomments_csv
```

```
%sql
select video_id,title,likes,dislikes from GBvideos_csv
where (likes > 1000) and (dislikes < 100)
```

```
%sql
select GBvideos_csv.video_id,GBvideos_csv.channel_title,GBvideos_csv.title,GBcomments_csv.comment_text from GBvideos_csv
join
GBcomments_csv
on
GBvideos_csv.video_id = GBcomments_csv.video_id
order by title desc

```

```
#3
%sql
create view vw_allvideos
as
select GBvideos_csv.video_id,GBvideos_csv.channel_title,GBvideos_csv.title from GBvideos_csv
union
select USvideos_csv.video_id,USvideos_csv.channel_title,USvideos_csv.title from USvideos_csv
```

```
%sql
select * from vw_allvideos
```

```
%sql
create view vw_allcomments
as
select GBcomments_csv.video_id,GBcomments_csv.comment_text,GBcomments_csv.likes,GBcomments_csv.replies from GBcomments_csv
union
select UScomments_csv.video_id,UScomments_csv.comment_text,UScomments_csv.likes,UScomments_csv.replies from UScomments_csv
```

```
%sql
select * from vw_allcomments
```

```
%sql
select video_id,count(likes) as total_likes,count(replies) as total_replies from GBcomments_csv
group by video_id,replies
```

```
%sql 
select distinct(channel_title),title,views from GBvideos_csv
where views > 10000000 
order by channel_title asc
```

```
%sql
select * from GBcomments_csv
where likes > 1000 and replies > 500
```

```
%sql
select channel_title,sum(views) as totalviews from GBvideos_csv
group by channel_title
```

```
%sql
select distinct(vw_allvideos.video_id),vw_allvideos.title,count(vw_allcomments.commnet_text) from vw_allvideos
join
vw_allcomments
on
vw_allcomments.video_id = vw_allvideos.video_id
group by vw_allcomments.video_id
```

```
%sql
select video_id,count(comment_text) as total_comments from vw_allcomments
group by video_id
having count(comment_text) > 500
```
